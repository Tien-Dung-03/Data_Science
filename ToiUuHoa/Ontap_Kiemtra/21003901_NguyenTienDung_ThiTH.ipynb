3. (2 điểm) Tìm learning rate theo phương pháp exact-line search trong thuật toán Gradient Descent của bài toán trên với điểm khởi tạo là $x^{(0)}=(1;0)$. Gọi điểm cập nhật trong trường hợp này là $x^{(k)}_{\text{exact}}$, vẽ đồ thị mô tả cho sai số $|f(x^{(k)}_{\text{exact}})-f(p^*)|$.

```python
import numpy as np
import math
```

```python
b = np.array([-1, 2])
A = np.array([[1, 0], [0, 3]])
```

```python
def objective_function(x):
    return 1/2 * x * A * x - b * x

def gradient(x):
    return 1/2 * A * x - b 

def gradient_descent(learning_rate, max_iterations, initial_x):
    x = [initial_x]
    for i in range(max_iterations):
        grad_x = gradient(x[-1])
        x_new = x[-1] - learning_rate * grad_x
        x.append(x_new) 
        if abs(gradient(x_new).all()) < 1e-5:
            break
    return x
```

```python
learning_rate = 0.9
max_iterations = 100
initial_x = 3
# x, it = gradient_descent(learning_rate, max_iterations, initial_x)

x = gradient_descent(learning_rate, max_iterations, initial_x)
print(f'x: ({x[-1]})')
print(f'Gia tri toi uu: {objective_function(x[-1])}')
```
